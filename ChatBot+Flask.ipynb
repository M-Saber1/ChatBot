```python
#Used in Tensorflow Model
import numpy as np
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Dropout



#import tflearn
import random

#Usde to for Contextualisation and Other NLP Tasks.
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download('stopwords')

#Other
import json
import string
import pickle
import warnings
warnings.filterwarnings("ignore")
nltk.download('punkt')
```

```python
print("Processing the Intents.....")
with open('DermaTection.json') as json_data:
    data = json.load(json_data)
```

```python
words = []
classes = []
data_x = []
data_y = []
print("Looping through the DermaTection Data to Convert them to words, classes, documents and ignore_words.......")
for intent in data['intents']:
    for pattern in intent['patterns']:
        # tokenize each word in the sentence
        tokens = nltk.word_tokenize(pattern)
        # add to our words list
        words.extend(tokens)
        data_x.append(pattern)
        # add to documents in our corpus
        data_y.append(intent['tag'])
        # add to our classes list
        if intent['tag'] not in classes:
            classes.append(intent['tag'])
```

```python
lemmatizer = WordNetLemmatizer()
print("Stemming, Lowering and Removing Duplicates.......")
words = [lemmatizer.lemmatize(word.lower()) for word in words if word not in string.punctuation]
words = sorted(set(words))

# remove duplicates
classes = sorted(set(classes))


print (len(classes), "classes", classes)
print (len(words), "unique stemmed words", words)
```

```python
print("Creating the Data for our Model.....")
training = []
out_empty = [0]*len(classes)
print("Creating an List (Empty) for Output.....")


print("Creating Traning Set, Bag of Words for our Model....")
for idx,doc in enumerate(data_x):
    # initialize our bag of words
    bow = []
    text=lemmatizer.lemmatize(doc.lower())
    # list of tokenized words for the pattern
    for word in words:
        bow.append(1) if word in text else bow.append(0)

    # output is a '0' for each tag and '1' for current tag
    output_row = list(out_empty)
    output_row[classes.index(data_y[idx])] = 1

    training.append([bow, output_row])

random.shuffle(training)
training = np.array(training,dtype=object)

train_x = np.array(list(training[:,0]))
train_y = np.array(list(training[:,1]))
```

```python
model = Sequential()
model.add(Dense(128,input_shape=(len(train_x[0]),),activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(64,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(17,input_shape=(len(train_y[0]),),activation="softmax"))
```

```python
adam = tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
model.compile(loss='categorical_crossentropy',optimizer=adam,metrics=["accuracy"])
```

```python
print(model.summary())
```

```python
model.fit(x= train_x ,y=train_y,epochs=150,verbose=1)
```

```python
def clean_text(text):
    # It Tokenize or Break it into the constituents parts of Sentense.
    tokens = nltk.word_tokenize(text)
    # Stemming means to find the root of the word.
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return tokens

# Return the Array of Bag of Words: True or False and 0 or 1 for each word of bag that exists in the Sentence
def bag_of_words(text, vocab):
    tokens = clean_text(text)
    bow = [0]*len(vocab)
    for w in tokens:
        for idx,word in enumerate(vocab):
            if word == w:
                bow[idx] = 1
    return np.array(bow)

def pred_class(text,vocab,labels):
    bow = bag_of_words(text,vocab)
    result = model.predict(np.array([bow]))[0]
    thresh = 0.5
    y_pred = [[indx,res] for indx,res in enumerate(result) if res > thresh]
    y_pred.sort(key=lambda x:x[1],reverse =True)
    return_list = []
    for r in y_pred:
        return_list.append(labels[r[0]])
    return return_list

def get_response(intents_list,intents_json):
    if len(intents_list) == 0:
        result = "Sorry I Dont Understant."
    else:
        tag = intents_list[0]
        list_of_intents = intents_json["intents"]
        for i in list_of_intents:
            if i["tag"] == tag:
                result = random.choice(i["responses"])
                break
        return result

```
